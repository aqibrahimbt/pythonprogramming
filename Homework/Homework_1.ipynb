### *For Loop Tasks*
 1.⁠ ⁠*Print Numbers*  
   Write a program that prints numbers from 1 to 10 using a ⁠ for ⁠ loop.

 2.⁠ ⁠*Sum of Numbers*  
   Use a ⁠ for ⁠ loop to calculate the sum of all numbers from 1 to 100.

 3.⁠ ⁠*List Iteration*  
   Given a list of fruits ⁠ ["apple", "banana", "cherry"] ⁠, print each fruit using a ⁠ for ⁠ loop.

 4.⁠ ⁠*Multiplication Table*  
   Write a program that generates the multiplication table of a number entered by the user (e.g., for 5: 5 × 1 = 5, 5 × 2 = 10, ...).

 5.⁠ ⁠*Reverse String*  
   Use a ⁠ for ⁠ loop to reverse a string entered by the user. For example, input: ⁠ "hello" ⁠, output: ⁠ "olleh" ⁠.
